```python
def transform(example_batch):
    # Take a list of PIL images and turn them to pixel values
    inputs = feature_extractor([x for x in example_batch['image']], return_tensors='pt')

    # Don't forget to include the labels!
    inputs['labels'] = example_batch['labels']
    return inputs
```

```python
prepared_ds = ds_final.with_transform(transform)
```

```python
prepared_ds[0:2]
```

```python
os.environ['CUDA_LAUNCH_BLOCKING'] = "1"
```

```python
import torch

def collate_fn(batch):
    return {
        'pixel_values': torch.stack([x['pixel_values'] for x in batch]),
        'labels': torch.tensor([x['labels'] for x in batch])
    }
```

```python
train_size = 1500#int(0.8 * len(ds_final))
test_size = len(prepared_ds) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(prepared_ds, [train_size, test_size])
train_dataset

train_final_size = 1200;
test_final_size = len(train_dataset) - train_final_size
train_final_dataset, test_final_dataset = torch.utils.data.random_split(train_dataset, [train_final_size, test_final_size]);
```

```python
import numpy as np
from datasets import load_metric

metric = load_metric("accuracy")
def compute_metrics(p):
    return metric.compute(predictions=np.argmax(p.predictions, axis=1), references=p.label_ids)
```

```python
from transformers import ViTForImageClassification

labels = [1, 2, 3, 4]

model = ViTForImageClassification.from_pretrained(
    vitext,
    num_labels=4,
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
  output_dir="./vit-base-cancer",
  per_device_train_batch_size=16,
  evaluation_strategy="steps",
  num_train_epochs=5,
  fp16=False,
  save_steps=100,
  eval_steps=100,
  logging_steps=10,
  learning_rate=2e-4,
  save_total_limit=2,
  remove_unused_columns=False,
  push_to_hub=False,
  report_to='tensorboard',
  load_best_model_at_end=True,
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=collate_fn,
    compute_metrics=compute_metrics,
    train_dataset=train_final_dataset,
    eval_dataset=test_final_dataset,
    tokenizer=feature_extractor,
)
```

```python
train_results = trainer.train()
trainer.save_model()
trainer.log_metrics("train", train_results.metrics)
trainer.save_metrics("train", train_results.metrics)
trainer.save_state()
```

```python
from datasets import load_metric

metric = load_metric("accuracy")
```

```python
metrics = trainer.evaluate(val_dataset)
trainer.log_metrics("eval", metrics)
trainer.save_metrics("eval", metrics)
```
